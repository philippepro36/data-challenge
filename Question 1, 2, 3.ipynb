```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Azuki = pd.read_csv("Azuki_3.csv", low_memory=False)
BAYC = pd.read_csv("BAYC_3.csv", low_memory=False)
MAYC = pd.read_csv("MAYC_3.csv", low_memory=False)
Moonbirds = pd.read_csv("Moonbirds_3.csv", low_memory=False)
Otherdeed = pd.read_csv("Otherdeed_3.csv", low_memory=False)
```

```python
Azuki
```

```python
Azuki_dates = Azuki[["timestamp"]]
Azuki_dates['Date'] = pd.to_datetime(Azuki_dates['timestamp'], format='%d/%m/%Y %H:%M')
Azuki_dates['Date'] = Azuki_dates['Date'].dt.strftime('%d/%m/%Y')
Azuki_dates = Azuki_dates[['Date']]
Azuki_2 = pd.concat([Azuki_dates, Azuki[["usd_price"]]], axis=1)
Azuki_2.index = Azuki_2["Date"]
Azuki_2.index = pd.to_datetime(Azuki_2.index, format="%d/%m/%Y")
Azuki_2
```

```python
BAYC_dates = BAYC[["timestamp"]]
BAYC_dates['Date'] = pd.to_datetime(BAYC_dates['timestamp'], format='%d/%m/%Y %H:%M')
BAYC_dates['Date'] = BAYC_dates['Date'].dt.strftime('%d/%m/%Y')
BAYC_dates = BAYC_dates[['Date']]
BAYC_2 = pd.concat([BAYC_dates, BAYC[["usd_price"]]], axis=1)
BAYC_2.index = BAYC_2["Date"]
BAYC_2.index = pd.to_datetime(BAYC_2.index, format="%d/%m/%Y")
BAYC_2
```

```python
MAYC_dates = MAYC[["timestamp"]]
MAYC_dates['Date'] = pd.to_datetime(MAYC_dates['timestamp'], format='%d/%m/%Y %H:%M')
MAYC_dates['Date'] = MAYC_dates['Date'].dt.strftime('%d/%m/%Y')
MAYC_dates = MAYC_dates[['Date']]
MAYC_2 = pd.concat([MAYC_dates, MAYC[["usd_price"]]], axis=1)
MAYC_2.index = MAYC_2["Date"]
MAYC_2.index = pd.to_datetime(MAYC_2.index, format="%d/%m/%Y")
MAYC_2
```

```python
Moonbirds_dates = Moonbirds[["timestamp"]]
Moonbirds_dates['Date'] = pd.to_datetime(Moonbirds_dates['timestamp'], format='%d/%m/%Y %H:%M')
Moonbirds_dates['Date'] = Moonbirds_dates['Date'].dt.strftime('%d/%m/%Y')
Moonbirds_dates = Moonbirds_dates[['Date']]
Moonbirds_2 = pd.concat([Moonbirds_dates, Moonbirds[["usd_price"]]], axis=1)
Moonbirds_2.index = Moonbirds_2["Date"]
Moonbirds_2.index = pd.to_datetime(Moonbirds_2.index, format="%d/%m/%Y")
Moonbirds_2
```

```python
Otherdeed_dates = Otherdeed[["timestamp"]]
Otherdeed_dates['Date'] = pd.to_datetime(Otherdeed_dates['timestamp'], format='%d/%m/%Y %H:%M')
Otherdeed_dates['Date'] = Otherdeed_dates['Date'].dt.strftime('%d/%m/%Y')
Otherdeed_dates = Otherdeed_dates[['Date']]
Otherdeed_2 = pd.concat([Otherdeed_dates, Otherdeed[["usd_price"]]], axis=1)
Otherdeed_2.index = Otherdeed_2["Date"]
Otherdeed_2.index = pd.to_datetime(Otherdeed_2.index, format="%d/%m/%Y")
Otherdeed_2
```

```python
Azuki_daily_transaction = Azuki_2.resample('d').count()
Azuki_daily_transaction = Azuki_daily_transaction['usd_price'].to_frame()
Azuki_daily_transaction = Azuki_daily_transaction.rename(columns={'usd_price': "daily transactions"})
Azuki_daily_transaction.plot(title="Number of daily transactions over time (Azuki)", figsize=(15,10), color="C3", ylabel="transactions")
```

```python
BAYC_daily_transaction = BAYC_2.resample('d').count()
BAYC_daily_transaction = BAYC_daily_transaction['usd_price'].to_frame()
BAYC_daily_transaction = BAYC_daily_transaction.rename(columns={'usd_price': "daily transactions"})
BAYC_daily_transaction.plot(title="Number of daily transactions over time (BAYC)", figsize=(15,10), color="C0", ylabel="transactions")
```

```python
MAYC_daily_transaction = MAYC_2.resample('d').count()
MAYC_daily_transaction = MAYC_daily_transaction['usd_price'].to_frame()
MAYC_daily_transaction = MAYC_daily_transaction.rename(columns={'usd_price': "daily transactions"})
MAYC_daily_transaction.plot(title="Number of daily transactions over time (MAYC)", figsize=(15,10), color="C2", ylabel="transactions")
```

```python
Moonbirds_daily_transaction = Moonbirds_2.resample('d').count()
Moonbirds_daily_transaction = Moonbirds_daily_transaction['usd_price'].to_frame()
Moonbirds_daily_transaction = Moonbirds_daily_transaction.rename(columns={'usd_price': "daily transactions"})
Moonbirds_daily_transaction.plot(title="Number of daily transactions over time (Moonbirds)", figsize=(15,10), color="C4", ylabel="transactions")
```

```python
Otherdeed_daily_transaction = Otherdeed_2.resample('d').count()
Otherdeed_daily_transaction = Otherdeed_daily_transaction['usd_price'].to_frame()
Otherdeed_daily_transaction = Otherdeed_daily_transaction.rename(columns={'usd_price': "daily transactions"})
Otherdeed_daily_transaction.plot(title="Number of daily transactions over time (Otherdeed)", figsize=(15,10), color="C5", ylabel="transactions")
```

```python
eth = pd.read_csv("ETH-USD.csv")   
eth = eth[['Date', 'Open']]
eth['Date'] = pd.to_datetime(eth['Date'], format='%Y-%m-%d')
eth['Date'] = eth['Date'].dt.strftime('%d/%m/%Y')
eth.index = eth["Date"]
eth.index = pd.to_datetime(eth.index, format="%d/%m/%Y")
eth = eth[["Open"]]
eth
```

```python
BAYC_daily_transaction
```

```python
Azuki_eth = pd.merge(Azuki_daily_transaction, eth, on='Date', how='left')
corr_matrix = Azuki_eth[['daily transactions', 'Open']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Reds')
plt.title('Heatmap of correlation between the number of daily transactions (Azuki) and the opening price of ETH')
```

```python
BAYC_eth = pd.merge(BAYC_daily_transaction, eth, on='Date', how='left')
corr_matrix = BAYC_eth[['daily transactions', 'Open']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
plt.title('Heatmap of correlation between the number of daily transactions (BAYC) and the opening price of ETH')
```

```python
MAYC_eth = pd.merge(MAYC_daily_transaction, eth, on='Date', how='left')
corr_matrix = MAYC_eth[['daily transactions', 'Open']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Greens')
plt.title('Heatmap of correlation between the number of daily transactions (MAYC) and the opening price of ETH')
```

```python
Moonbirds_eth = pd.merge(Moonbirds_daily_transaction, eth, on='Date', how='left')
corr_matrix = Moonbirds_eth[['daily transactions', 'Open']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Purples')
plt.title('Heatmap of correlation between the number of daily transactions (Moonbirds) and the opening price of ETH')
```

```python
Otherdeed_eth = pd.merge(Otherdeed_daily_transaction, eth, on='Date', how='left')
corr_matrix = Otherdeed_eth[['daily transactions', 'Open']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Oranges')
plt.title('Heatmap of correlation between the number of daily transactions (Otherdeed) and the opening price of ETH')
```
